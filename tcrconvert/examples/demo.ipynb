# Example converting 10X data to Adaptive format

This notebook demonstrates how to convert 10X T-cell receptor gene names to the Adaptive format.

Import libraries and define filepaths

```python
import pandas as pd
from tcrconvert import convert

dat10x_f = '/Users/emmabishop/workspace/tcrconvert/tcrconvert/data/example_10x.csv'
```

Load example data

```python
dat10x = pd.read_csv(dat10x_f)
dat10x
```

Take a selection of columns and add a row of nonsense genes, to demonstrate what happens when you data isn't in the IMGT reference.

```python
my_df = dat10x[['barcode', 'is_cell', 'v_gene', 'j_gene', 'cdr3']].copy()
my_df.loc[5] = ['BARCODE', 'True', 'TRBVXYZ', 'TRBJ555', 'CDR3HERE']
my_df
```

Do the conversion

```python
out = convert.convert_tcr(my_df, fmt_from='tenx', fmt_to='adaptive', convert_cols=True)
out
```
